```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xArray = np.linspace(0,2*np.pi, 100)

def function(x):
    a = 4.0
    b = 3.0
    yAnswer = a * np.exp(-b * x)
    return yAnswer
```

```python
yData = function(xArray)
plt.plot(xArray, yData)
plt.xlabel("x")
plt.ylabel("4e^(-3x)")
plt.savefig("Journal2-Prompt1 Graph", bbox_inches = "tight", dpi = 400)
```
