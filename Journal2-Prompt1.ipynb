```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xArray = np.linspace(0,2*np.pi, 100)

def function(x):
    a = 1.0
    b = 1.0
    yAnswer = a * np.exp(-b * x)
    return yAnswer
```
