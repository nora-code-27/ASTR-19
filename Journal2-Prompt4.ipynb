```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
list = np.random.triangular(left = 0, mode = 30, right = 100, size = 1000)
#print(list)
```

```python
graph = plt.hist(list, bins = 100)
#graph = plt.plot(list, c = "r")
plt.ylim = (100)
plt.xlabel("Random Variable")
plt.ylabel("Probability Dencity")
plt.savefig("Journal2-Prompt4 Graph", bbox_inches = "tight", dpi = 400)
```
